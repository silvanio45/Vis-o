```python
import cv2
```

```python
Imagem = cv2.imread('/content/pessoas.jpg')
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(Imagem)
```

```python
detector = cv2.CascadeClassifier('/content/fullbody.xml')
```

```python
Img_cinza = cv2.cvtColor(Imagem, cv2.COLOR_BGR2GRAY)
cv2_imshow(Img_cinza)
```

```python
deteccoes = detector.detectMultiScale(Img_cinza)
```

```python
for (x, y, l, a) in deteccoes:
  cv2.rectangle(Imagem, (x,y), (x + l, y + a), (0, 255, 0) ,2)
```
